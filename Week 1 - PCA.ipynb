```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
means = [2, 2]
cov = [[1, 0], [0, 1]]
N = 50
X = np.random.multivariate_normal(means,cov,N)
```

```python
fig, axes = plt.subplots()
axes.scatter(X[:,0], X[:,1], c="green")
plt.show()
```

```python
def pca_im(X):
    X = X - X.mean(axis=0)
    X_std = X/ np.std(X, axis=0)
    cov_mat = np.cov(X_std.T)
    e, v = np.linalg.eig(cov_mat)
    print('Eigenvalue:', e)
    print('Eigenvector', v)
    max_vec = v[:,e.argmax()]
    result = X_std.dot(max_vec) * max_vec[:, np.newaxis]
    return result
```

```python
dt = pca_im(X)
```

```python
plt.scatter(dt[0], dt[1], c="green")
plt.show()
```

```python
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
X = df.loc[:,features].values
```

```python
fig, axes = plt.subplots()
axes.scatter(X[:,0], X[:,1], c="green")
plt.show()
```

```python
def pca_im_2(X):
    X = X - X.mean(axis=0)
    X_std = X/ np.std(X, axis=0)
    cov_mat = np.cov(X_std.T)
    e, v = np.linalg.eig(cov_mat)
    print('Eigenvalue:', e)
    print('Eigenvector', v)
    max_vec = v[:, e.argsort()[::-1][:2]]
    result = X_std.dot(max_vec)
    return result
```

```python
dt_2 = pca_im_2(X)
```

```python
plt.scatter(dt_2[:,0], dt_2[:,1], c="green")
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
x = df.loc[:, features].values
```

```python
fig, axes = plt.subplots()
axes.scatter(x[:,0], x[:,1], c="green")
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
```

```python
plt.scatter(principalComponents[:,0], principalComponents[:,1], c='green')
plt.show()
```

```python
plt.plot(principalComponents[:,0], principalComponents[:,1], color='black', marker='o', markersize = 5)
plt.axvline(x=0, ymin=0, ymax=1, lw=1.5, label='boudary', c='red')
plt.axhline(y=0, xmin=0, xmax=1, lw=1.5, c='red')
plt.show()
```
